```python
import pandas as pd
```

```python
nba_4factor_path = "Resources/sportsref_fourfactors.xlsx"
nba_team_path = "Resources/sportsref_team_21.xlsx"
nba_opp_path = "Resources/sportsref_opp_21.xlsx"
```

```python
nba_team = pd.read_excel(nba_team_path,"Worksheet")
nba_opp = pd.read_excel(nba_opp_path,"Worksheet")
nba_4fac = pd.read_excel(nba_4factor_path,"Worksheet")
```

```python
#Offensive field goal % = highest should be rank 1
#Defensive field goal % = Lowest should be rank 1
#Offensive turnover % = lowest should be rank 1
#Defensive turnover % = highest should be rank 1
#Offensive free throw ratio = highest should be rank 1
#Defensive free throw ratio = lowest should be rank 1
#Offensive rebounding % = highest should be rank 1
#Defensive rebounding % = highest should be rank 1
```

```python
nba_4fac["R(OeFG%)"] = nba_4fac["OeFG%"].rank(ascending = 0)
nba_4fac["R(DeFG%)"] = nba_4fac["DeFG%"].rank(ascending = 1)
nba_4fac["R(OTOV%)"] = nba_4fac["OTOV%"].rank(ascending = 1)
nba_4fac["R(DTOV%)"] = nba_4fac["DTOV%"].rank(ascending = 0)
nba_4fac["R(OFT/FGA)"] = nba_4fac["OFT/FGA"].rank(ascending = 1)
nba_4fac["R(DFT/FGA)"] = nba_4fac["DFT/FGA"].rank(ascending = 0)
nba_4fac["R(ORB%)"] = nba_4fac["ORB%"].rank(ascending = 0)
nba_4fac["R(DRB%)"] = nba_4fac["DRB%"].rank(ascending = 0)
```

```python
nba_pytha = nba_team.merge(nba_opp,on="Team",suffixes=('_team','_opp'))
```

```python
nba_pytha = nba_pytha[["Team","W_team","L_team","PTS_team","PTS_opp"]].sort_values("Team")
```
